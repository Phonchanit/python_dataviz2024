```python
import pandas as pd
```

```python
# เริ่มจากการตรวจสอบ ข้อมูลคดี (crime_scene_report.csv)
crime_scene_report = pd.read_csv('crime_scene_report.csv')
crime_scene_report.head()
```

```python
interview = pd.read_csv('interview.csv')
interview.head()
```

```python
# กรองข้อมูลคดีฆาตกรรมที่ SQL City วันที่ 15 มกราคม 2018
crime_sql_city = crime_scene_report[(crime_scene_report['city'] == 'SQL City') &
                                    (crime_scene_report['date'] == 20180115) &
                                    (crime_scene_report['type'] == 'murder')]
crime_sql_city
```

```python
# แปลงคอลัมน์ 'date' ให้เป็นชนิด string เพื่อใช้ในการกรองข้อมูล
crime_scene_report['date'] = crime_scene_report['date'].astype(str)
```

```python
# กรองคดีฆาตกรรมที่เกิดขึ้นใน SQL City ในวันที่ 15 มกราคม 2018
murder_case = crime_scene_report.query("date == '20180115' and type == 'murder' and city == 'SQL City'")
murder_case
```

```python
person = pd.read_csv('person.csv')
person.head()
```

```python
# ค้นหาผู้ที่อาศัยอยู่บนถนน Northwestern Dr
northwestern_residents = person[person['address_street_name'] == 'Northwestern Dr']
```

```python
# ค้นหาบ้านเลขที่สูงสุดบนถนน Northwestern Dr
top_address = northwestern_residents.loc[northwestern_residents['address_number'].idxmax()]
top_address
```

```python
# แยกคอลัมน์ชื่อ 'name' ออกเป็น 'first_name' และ 'last_name'
person[['first_name', 'last_name']] = person['name'].str.split(' ', expand=True)
```

```python
# กรองข้อมูลเพื่อหาบุคคลที่ชื่อ Annabel และอยู่บนถนน Franklin Ave
annabel_resident = person.query("first_name == 'Annabel' and address_street_name == 'Franklin Ave'")
annabel_resident
```
