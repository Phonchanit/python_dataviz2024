# HW6
บรรยายเกี่ยวกับผู้กำกับชั้นนำของอินเดีย 3 คน ว่าทำไมเราถึงคิดว่าเป็นผู้กำกับชั้นนำ และชั้นนำยังไง

````
# prompt: วิเคราะห์ว่าผู้กำกับ 3 คนไหนที่เป็นผู้กำกับชั้นนำโดยอ้างอิงจาก Rating หนังที่ผู้กำกับแต่ละคนกำกับ

# นับจำนวนหนังของแต่ละ Director
director_counts = movie_df['Director'].value_counts()
````

````
# เลือก Top 3 Directors ที่กำกับหนังเยอะที่สุด
top_3_directors = director_counts.head(3).index.tolist()
````

````
# สร้าง DataFrame ว่างเพื่อเก็บผลลัพธ์
top_directors_analysis = pd.DataFrame(columns=['Director', 'Mean Rating', 'Num Movies'])
````

````
# วนลูปคำนวณค่าเฉลี่ย Rating และจำนวนหนังของ Top 3 Directors
for director in top_3_directors:
    director_movies = movie_df[movie_df['Director'] == director]
    mean_rating = director_movies['Rating'].mean()
    num_movies = director_movies.shape[0]
    # Use pd.concat instead of .append
    top_directors_analysis = pd.concat([top_directors_analysis, pd.DataFrame({'Director': [director], 'Mean Rating': [mean_rating], 'Num Movies': [num_movies]})], ignore_index=True)
````

````

# เรียงลำดับ DataFrame ตาม Mean Rating จากมากไปน้อย
top_directors_analysis = top_directors_analysis.sort_values(by='Mean Rating', ascending=False)

# แสดงผลลัพธ์
print(top_directors_analysis)
````

 สามารถวิเคราะห์ได้ว่าผู้กำกับ 3 คนที่เป็นผู้กำกับชั้นนำ (ตาม Mean Rating) คือ:
1. Jayant Desai (ผู้กำกับที่มี Mean Rating สูงสุด)
