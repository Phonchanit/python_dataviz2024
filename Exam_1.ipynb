```python
TestData_path = 'test-TidyData_2024.csv'  # Replace with your file path
data = pd.read_csv(TestData_path)
```

```python
#1: ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019
data['year'] = data['month'].apply(lambda x: int(x.split('_')[-1]))

# Filter data for years 2018 and 2019
data_2018 = data[data['year'] == 2018]
data_2019 = data[data['year'] == 2019]

# Calculate the mean Chlo_a for each station in 2018 and 2019
mean_chlo_a_2018 = data_2018.groupby('station')['Chlo_a'].mean()
mean_chlo_a_2019 = data_2019.groupby('station')['Chlo_a'].mean()

# Combine the results into a single DataFrame for comparison
chlo_a_comparison = pd.DataFrame({
    'Mean_Chlo_a_2018': mean_chlo_a_2018,
    'Mean_Chlo_a_2019': mean_chlo_a_2019
}).reset_index()

# Display the comparison
print(chlo_a_comparison)
```

```python
#2: สร้างตารางค่าเฉลี่ย Chlo_b ทุกๆ station ของแต่ละเดือน
# Extract month name from the 'month' column
data['month_name'] = data['month'].apply(lambda x: x.split('_')[0])

# Calculate the mean Chlo_b for each station for each month
mean_chlo_b_monthly = data.groupby(['station', 'month_name'])['Chlo_b'].mean().unstack()

# Display the table
print(mean_chlo_b_monthly)
```
