# **Data**
show_id: รหัสเฉพาะของรายการหรือภาพยนตร์แต่ละรายการ

```python
# prompt: แบ่งข้อมูลจาก column ชื่อ director ออกมาโดยเอามาแค่ข้อมูลที่คั่นด้วย ','  เก็บ show_id และ title นำไปเก็บใน DataFrame ชื่อ directers

import pandas as pd

# Assuming df is your DataFrame

# Create an empty list to store the extracted data
directors_data = []

# Iterate through the DataFrame rows
for index, row in df.iterrows():
  director_list = row['director'].split(',')
  for director in director_list:
    directors_data.append({'show_id': row['show_id'], 'title': row['title'], 'director': director.strip()})

# Create a new DataFrame from the extracted data
directors = pd.DataFrame(directors_data)

# Display the new DataFrame
print(directors)
```

```python
# prompt: ทำ DataFrame ชื่อ directors เป็นไฟล์ .csv

# Save the DataFrame to a CSV file
directors.to_csv('directors.csv', encoding='utf-8', index=False)

# Download the CSV file
files.download('directors.csv')
```

```python
# prompt: สร้าง column ใหม่ชื่อ country1 เก็บข้อมูลมาแค่ประเทศแรกจากคอลัมน์ country

# สร้าง column ใหม่ชื่อ country1 เก็บข้อมูลมาแค่ประเทศแรกจากคอลัมน์ country
df['country1'] = df['country'].str.split(',').str[0]

# แสดงผลลัพธ์
print(df[['country', 'country1']])
```

```python
# prompt: เพิ่ม column ใหม่ชื่อ date_added_1 เก็บข้อมูลจาก date_added มาทำเป็น datetime ในรูปแบบวันเดือนปี

# Convert 'date_added' to datetime objects
df['date_added_1'] = pd.to_datetime(df['date_added'], errors='coerce')

# Extract year, month, and day from the datetime objects
df['date_added_1'] = df['date_added_1'].dt.strftime('%Y-%m-%d')

print(df[['date_added', 'date_added_1']])
```

```python
# prompt: เพิ่ม TV- หน้า G, หน้า PG, R, NR, UR

# Assuming df is your DataFrame

# Create a dictionary to map existing ratings to new ratings
rating_mapping = {
    'G': 'TV-G',
    'PG': 'TV-PG',
    'R': 'TV-R',
    'NR': 'TV-NR',
    'UR': 'TV-UR'
}

# Apply the mapping to the 'rating' column
df['rating'] = df['rating'].replace(rating_mapping)

# Display the updated DataFrame
print(df)

# Save the DataFrame to a CSV file
df.to_csv('latest_data_with_updated_rating.csv', encoding='utf-8', index=False)

# Download the CSV file
files.download('latest_data_with_updated_rating.csv')

```

```python
# prompt: แสดงประเภทหนังหรือรายการที่นิยมมากที่สุดในแต่ละปี

# Group the data by release year and type, then count the number of occurrences for each type within each year.
popular_types_by_year = df.groupby(['release_year', 'type'])['show_id'].count().unstack()

# Find the most popular type for each year.
most_popular_type_by_year = popular_types_by_year.idxmax(axis=1)

# Display the result
print(most_popular_type_by_year)
```
